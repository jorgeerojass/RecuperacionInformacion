```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Configuraciones
headers = {"User-Agent": "Mozilla/5.0"}
base_url = "https://www.allrecipes.com/search"
query = "chicken"
max_recipes = 99
step = 24  # cantidad de resultados por pÃ¡gina

recipe_urls = set()
offset = 0

while len(recipe_urls) < max_recipes:
    # Construir URL con offset
    url = f"{base_url}?q={query}&offset={offset}"
    print(f"ğŸ” Buscando en: {url}")
    
    # Obtener y parsear HTML
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    links = soup.find_all("a", href=True)
    
    # Extraer URLs vÃ¡lidas de receta
    found = 0
    for link in links:
        href = link['href']
        full_url = urljoin(url, href)
        if "/recipe/" in full_url and full_url.startswith("https://www.allrecipes.com/recipe/"):
            if full_url not in recipe_urls:
                recipe_urls.add(full_url)
                found += 1
        if len(recipe_urls) >= max_recipes:
            break

    if found == 0:
        print("âš ï¸ No se encontraron mÃ¡s recetas. Fin.")
        break

    offset += step

# Mostrar resultados
print("\nâœ… Recetas encontradas:")
for recipe in sorted(recipe_urls):
    print(recipe)

print(f"\nğŸ”¢ Total: {len(recipe_urls)} recetas")
```

```python
import os
import time

# Ruta donde guardarÃ¡s los .html
save_dir = r"\data"
os.makedirs(save_dir, exist_ok=True)

# Guardar cada receta en un archivo HTML
for i, url in enumerate(sorted(recipe_urls)):
    try:
        print(f"ğŸ’¾ Guardando receta {i+1}: {url}")
        response = requests.get(url, headers=headers, timeout=10)
        filename = os.path.join(save_dir, f"receta_{i+1}.html")
        with open(filename, "w", encoding="utf-8") as f:
            f.write(response.text)
        time.sleep(1.5)  # Evitar bloqueo
    except Exception as e:
        print(f"âŒ Error al guardar {url}: {e}")
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Directorio donde estÃ¡n los archivos
html_dir = r"\data"

# Lista para guardar la info de cada receta
recetas_data = []

# Recorrer los archivos HTML
for i in range(1, 100):  # puedes ajustar el rango segÃºn cuÃ¡ntos guardaste
    try:
        file_path = os.path.join(html_dir, f"receta_{i}.html")
        with open(file_path, "r", encoding="utf-8") as file:
            soup = BeautifulSoup(file.read(), 'html.parser')

        # TÃ­tulo
        title = soup.find("meta", {'property': 'og:title'})
        title = title["content"] if title else "N/A"

        # Rating
        rating_section = soup.find("div", class_="mm-recipes-review-bar__rating")
        rating = rating_section.text.strip() if rating_section else "N/A"

        # Ingredientes
        ingredients = soup.find_all("li", class_="mm-recipes-structured-ingredients__list-item")
        ingredients_list = [i.text.strip() for i in ingredients]

        # Porciones y tiempo total
        servings = ""
        total_time = ""
        for s in soup.find_all("div", class_="mm-recipes-details__item"):
            if "Servings" in s.text:
                servings = s.find("div", class_="mm-recipes-details__value").text.strip()
            if "Total Time" in s.text:
                total_time = s.find("div", class_="mm-recipes-details__value").text.strip()

        # DescripciÃ³n
        description_tags = soup.find_all("p", class_="article-subheading text-utility-300")
        description = " ".join([d.text.strip() for d in description_tags])

        # Pasos
        steps = soup.find_all("li", class_="mntl-sc-block-group--LI")
        steps_list = []
        for step in steps:
            paragraph = step.find("p")
            if paragraph:
                steps_list.append(paragraph.text.strip())

        # NutriciÃ³n
        nutrition_rows = soup.find_all("tr", class_="mm-recipes-nutrition-facts-summary__table-row")
        nutrition = {}
        for row in nutrition_rows:
            label = row.find("td", class_="text-body-100")
            value = row.find("td", class_="text-body-100-prominent")
            if label and value:
                nutrition[label.text.strip()] = value.text.strip()

        # Agregar a la lista
        recetas_data.append({
            "TÃ­tulo": title,
            "Rating": rating,
            "Porciones": servings,
            "Tiempo total": total_time,
            "Ingredientes": ingredients_list,
            "DescripciÃ³n": description,
            "Pasos": steps_list,
            "NutriciÃ³n": nutrition
        })

    except Exception as e:
        print(f"âŒ Error en receta {i}: {e}")
```

```python
# Convertir a DataFrame y guardar
df = pd.DataFrame(recetas_data)
df.to_csv("recetas_extraidas.csv", index=False)
print("âœ… CSV creado con Ã©xito.")
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

# 1. Cargar el CSV y preparar el contexto
df = pd.read_csv('recetas_extraidas.csv')
df['contexto'] = (
    "TÃ­tulo: " + df['TÃ­tulo'].astype(str) + ". " +
    "Rating: " + df['Rating'].astype(str) + ". " +
    "Porciones: " + df['Porciones'].astype(str) + ". " +
    "Tiempo total: " + df['Tiempo total'].astype(str) + ". " +
    "Ingredientes: " + df['Ingredientes'].astype(str) + ". " +
    "DescripciÃ³n: " + df['DescripciÃ³n'].astype(str) + ". " +
    "Pasos: " + df['Pasos'].astype(str) + ". " +
    "NutriciÃ³n: " + df['NutriciÃ³n'].astype(str) + "."
)

# 2. Obtener embeddings de cada receta
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['contexto'].tolist())
embeddings = np.array(embeddings).astype('float32')

# 3. Indexar con FAISS
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)

# 4. Codificar la query y buscar las recetas mÃ¡s similares
query = "healthy chicken salad"  # Cambia aquÃ­ tu consulta
query_emb = model.encode([query]).astype('float32')
k = 5
distances, indices = index.search(query_emb, k)

# 5. Mostrar los resultados en un DataFrame
similarity_df = pd.DataFrame({
    'index': indices[0],
    'distance': distances[0]
})
similarity_df['contexto'] = df.iloc[similarity_df['index']]['contexto'].values
similarity_df = similarity_df.sort_values(by='distance')
similarity_df
```

```python
# Une el contexto de las recetas mÃ¡s similares en un solo string
contexto_similar = "\n\n".join(similarity_df['contexto'].tolist())
print(contexto_similar)
```
