CONTAR CUANTAS PALABRAS SIN REPETICION Y SIN TILDES HAY EN EL CORPUS

```python
import unicodedata, string

def contar_palabras(filepath):
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            texto = f.read()
            palabras = texto.split()
            normalizadas = {
                ''.join(c for c in unicodedata.normalize('NFD', p.lower().strip(string.punctuation))
                        if unicodedata.category(c) != 'Mn')
                for p in palabras if p.strip(string.punctuation)
            }
            return len(normalizadas)
    except FileNotFoundError:
        print(f"Archivo no encontrado: {filepath}")
    except Exception as e:
        print(f"Error: {e}")

print("Palabras únicas:", contar_palabras('01_corpus_turismo.txt'))
```

REALIZAR LA TABLA MATRIZ TEERMINO-DOCUMENTO Y CALCULAR IDF

```python
import math

def normalizar(palabra):
    palabra = palabra.lower()
    acentos = str.maketrans('áéíóúüñ', 'aeiouun')
    return palabra.translate(acentos).strip('.,;:!?()"\'«»')

with open('01_corpus_turismo.txt', 'r', encoding='utf-8') as f:
    lineas = f.readlines()

documentos = [f'doc{i+1}' for i in range(len(lineas))]
frecuencias = {}
apariciones = {}
vocabulario = set()

for i, linea in enumerate(lineas):
    palabras = [normalizar(p) for p in linea.split()]
    doc = documentos[i]
    conteo = {}
    for p in palabras:
        if p:
            conteo[p] = conteo.get(p, 0) + 1
    frecuencias[doc] = conteo
    for p in conteo:
        vocabulario.add(p)
        apariciones.setdefault(p, set()).add(doc)

N = len(documentos)
cabecera = ['Palabra'] + documentos + ['IDF']
print('\t'.join(cabecera))

for palabra in sorted(vocabulario):
    fila = [palabra]
    df = len(apariciones[palabra])
    for doc in documentos:
        fila.append(str(frecuencias[doc].get(palabra, 0)))
    idf = round(math.log10(N / df), 5) if df else 0
    fila.append(str(idf))
    print('\t'.join(fila))
```
